```python
import pandas as pd
```

```python
data = pd.read_excel('HRData.xlsx')
df = pd.DataFrame(data)
df.head()
```

```python
df.columns
```

```python
# Renamed columns by removing the spaces contained in original names

new_column_names = {' EmployeeName': 'EmployeeName', ' Department': 'Department', ' Position': 'Position',
                    ' Salary': 'Salary', ' EmploymentType':'EmploymentType', ' HireDate': 'HireDate',
                    ' ResignationDate': 'ResignationDate', ' BirthDate': 'BirthDate', ' Gender':'Gender',
                   ' MaritalStatus': 'MaritalStatus', ' emp_performance_ratings': 'emp_performance_ratings',
                   ' attrition': 'attrition', ' Monthname': 'Monthname', ' DayName': 'DayName',
                    ' EmployeeTenure': 'EmployeeTenure', ' EmpAge ': 'EmpAge'}

df.rename(columns=new_column_names, inplace=True)
```

```python
df.columns=df.columns.str.strip() #to remove all spaces once
```

```python
df.info()
```

```python
df['EmployeeID'] = df['EmployeeID'].astype(str)
```

```python
# checked for null value

df.isnull().sum()
```

```python
# addressing null values and populating empty columns
```

```python
# for ResignationDate column

# Duplicated the ResignationDate column as a new column
df['Resignation_date'] = df['ResignationDate'].copy()

# Replaced "NaT" with today's date in the new column
replacement_date = pd.to_datetime('2024-05-09')

# Replace NaT with the specified date
df['Resignation_date'] = df['Resignation_date'].fillna(replacement_date)
```

```python
# viewed the duplicated column and value replacement 

vieww = df[['ResignationDate', 'Resignation_date']]
vieww
```

```python
# for attrition column

df['attrition'] = df['ResignationDate'].notna()
df['attrition']
```

```python
# for EmployeeTenure column

df['EmployeeTenure'] = pd.to_timedelta(df['Resignation_date'] - df['HireDate']).dt.total_seconds() / (3600 * 24 * 365.25)
df['EmployeeTenure'] = df['EmployeeTenure'].astype(int)
df['EmployeeTenure']
```

```python
# for EmpAge column

date = pd.to_datetime("2024-05-09")

today = date.today()

def calculate_age(birthdate):
  if not pd.isna(birthdate):
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
  else:
    age = None 
  return age

df['c'] = df['BirthDate'].apply(calculate_age)
df['EmpAge']
```

```python
# for Monthname

df['Monthname'] = df['ResignationDate'].dt.month_name()
df['Monthname']
```

```python
# for DayName

df['DayName'] = df['ResignationDate'].dt.day_name()
df['DayName']
```

```python
import numpy as np

# Assuming your DataFrame is called 'df' and has a column named 'ratings'

# Define your rating scale (adjust the ranges and words as needed)
ratings_dict = {
    1: "Bad",
    2: "Okay",
    3: "Good",
    4: "Very Good",
    5: "Excellent"
}

# Sampled ratings as a Series 
ratings = df['emp_performance_ratings']  

# Converted ratings to words using vectorization
words = np.vectorize(ratings_dict.get)(ratings)

# Added a new column named Employee_Rating' with the converted words
df['Employee_Rating'] = words
df['Employee_Rating']
```

```python
# created a new column, 'Employment_Status' from attrition column

employment_status_map = {True: 'Resigned', False: 'Active'}

# Create a new column 'Employment_Status' based on the mapping
df['Employment_Status'] = df['attrition'].map(employment_status_map)
df['Employment_Status']
```

```python
df['EmpAge'].unique()
```

```python

# Defined age group ranges
age_groups = {
    '20-29': (20, 29),
    '30-39': (30, 39)
}

# Created a new 'Age_Group' column based on age ranges
def get_age_group(age):
  for group, (min_age, max_age) in age_groups.items():
    if min_age <= age <= max_age:
      return group
  return None 

df['Age_Group'] = df['c'].apply(get_age_group)
df['Age_Group']
```
